```python
import pandas as pd

df = pd.read_csv("loja.csv", sep=";", decimal=",")
df
```

```python
# ajustando a data
df["Data"] = pd.to_datetime(df["Data"])
df = df.sort_values("Data")
df
```

```python
# criando coluna para o mês
meses = df["Data"].apply(lambda x: str(x.year) + "-" + str(x.month))

# inserindo a coluna meses no dataframe, ao lado da data completa
df.insert(11, "Mes", meses)
df
```

```python
# filtrando pela coluna meses no dataframe
mes = df["Mes"].unique()

df_filtrado = df[df["Mes"] == mes[0]]
df_filtrado
```

```python
total_bairro = df_filtrado.groupby("Bairro")[["Total"]].sum().reset_index()
total_bairro
```

```python
notas = df_filtrado.groupby("Bairro")[["Nota"]].mean().reset_index()
notas
```

```python
produtos = df_filtrado.groupby("Tipo de produto")[["Quantidade"]].sum().reset_index()
produtos
```
