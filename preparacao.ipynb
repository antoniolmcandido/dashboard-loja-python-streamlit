```python
import pandas as pd

df = pd.read_csv("loja.csv", sep=";", decimal=",")
df
```

```python
# ajustando a data
df["Data"] = pd.to_datetime(df["Data"])
df = df.sort_values("Data")
df
```

```python
# caputurando o mês
df["Data"].apply(lambda x: str(x.month) + '/' + str(x.year))   
```
